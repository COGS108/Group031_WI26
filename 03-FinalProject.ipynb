## Abstract

Travel demands are a unique feature of professional sports schedules and may influence team performance through fatigue, sleep disruption, and circadian misalignment. This study investigates whether cumulative travel stress—measured through total travel distance, time-zone shifts, rest intervals, and back-to-back games—predicts declines in NBA teams’ Offensive Efficiency (OE) relative to their season averages. Using game-level data, we constructed multiple linear regression models to evaluate the relationship between travel-related variables and offensive performance metrics including Field Goal Percentage (FG%), Effective Field Goal Percentage (eFG%), and Turnover Percentage (TOV%). The models included fixed effects for team, opponent, and season to control for baseline differences in team strength.

Results indicate that playing on the road and competing in back-to-back games are the most statistically significant predictors of reduced offensive efficiency. Away teams experienced an average decrease of approximately 1.73 points per 100 possessions, while back-to-back games produced an additional decline of roughly 0.99 points. Contrary to the initial hypothesis, travel distance and number of time zones crossed did not significantly affect shooting efficiency metrics such as FG% or eFG%. However, measures related to accumulated travel—particularly time-zone crossings and travel over the previous three games—were associated with increases in turnover percentage, suggesting a potential cognitive fatigue effect.

Overall, the model explained a modest portion of offensive performance variance (R² ≈ 0.079), reflecting the complex and multifactorial nature of basketball outcomes. These findings suggest that while travel logistics alone may not directly reduce shooting accuracy, schedule compression and accumulated travel may still influence performance through increased mental fatigue and decision-making errors. Future research incorporating player-level data and more detailed fatigue metrics may help further clarify the relationship between travel stress and athletic performance.

## Research Question

Does cumulative travel stress, quantified by the interaction between total distance traveled (miles), time-zone shifts, and rest intervals (e.g., back to back games), predict a statistically significant decline in a team's Offensive Efficiency (OE) relative to their season average?

Specifically, can we develop a multiple linear regression model that accurately identifies the amount of fatigue tax on shooting percentages (teams FG%) and turnover rates for visiting teams?

This project will test whether geographic factors like eastward travel and high-mileage road trips correlate with specific performance drops. By isolating these variables, we aim to determine if a predictive threshold exists where travel fatigue becomes a dominant predictor of offensive variance.

## Background and Prior Work

The modern landscape of professional basketball is defined by a rigorous competitive structure that frequently pushes the boundaries of human physiological and cognitive limits.

Within the National Basketball Association (NBA), the standard 82-game regular season involves a high density of competition interspersed with frequent transmeridian travel, creating a unique environment where athletic performance is perpetually modulated by recovery status and circadian alignment.

Travel stress in this context is characterized by two distinct but overlapping phenomena: travel fatigue—the acute exhaustion from the logistics of movement—and jet lag, a circadian rhythm disorder caused by crossing multiple time zones. This biological asymmetry is critical, as the human circadian system is naturally more adept at adapting to westward travel (phase delay) than shortening its cycle through eastward travel (phase advance) <a name="cite_note-1"></a><sup>1</sup>.

Research indicates that sleep disruption is the primary mechanism through which travel stress erodes performance. On the first night after travel, athletes experience a predictable reduction in sleep duration based on the number of time zones crossed, with eastward travel being significantly more disruptive (averaging -24.5 minutes) than westward travel <a name="cite_note-2"></a><sup>2</sup>.

These disruptions can manifest as significant decrements in Offensive Efficiency (OE), a holistic metric that measures scoring effectiveness per possession. Specifically, studies have shown that eastward jet lag is associated with a 1.2% decrease in Effective Field Goal Percentage ($eFG\%$) differential <a name="cite_note-3"></a><sup>3</sup>.

Furthermore, mental fatigue disrupts psychomotor vigilance, essential for shooting accuracy and ball security, often leading to a rise in turnover rates (TOV%) at the end of long road trips.

Existing literature has also identified specific "tipping points" where travel fatigue becomes the dominant predictor of performance variance. One of the most robust findings is that cumulative time zone changes of three or more within a three-day period are significantly detrimental to performance <a name="cite_note-4"></a><sup>4</sup>.

Additionally, the back-to-back game scenario remains the most consistent predictor of decline; visiting teams playing on the second night of a back-to-back win only about 36% of the time.

While total mileage is a factor, its impact is most severe when combined with the duration of a multi-city tour, as the cumulative stress of changing environments leads to a measurable decline in $eFG\%$.

To capture these complex, non-linear interactions, modern sports analytics projects have increasingly moved from simple linear models toward advanced techniques like Random Forests and Gradient Boosting (LightGBM) to accurately identify and predict the "fatigue tax" <a name="cite_note-5"></a><sup>5</sup>.

## Hypothesis


We hypothesize that **cumulative travel stress**, more specifically the interaction of eastward travel (time zone shifts), total distance covered, and limited rest (playing back to back games), will serve as a **statistically significant predictor** of a decline in a visiting team’s Offensive Efficiency (OE).

### Specific Predictions
* **Directional Asymmetry:** Eastward travel across two or more time zones will result in a much more noticeable decline in Effective Field Goal Percentage ($eFG\%$) than westward travel of equal distance.
* **The "Fatigue Tax" Tipping Point:** The decline will follow a non-linear trend; the combination of high-mileage road trips and back-to-back scenarios creates a "tipping point" that significantly increases Turnover Rates ($TOV\%$).

### Reasoning
This prediction is based on the biological reality of **circadian asymmetry**, where the human body struggles more with "phase advance" (losing time traveling east) than "phase delay" (traveling west).



As highlighted in our background research, this sleep disruption impairs both the fine motor skills required for shooting accuracy and the cognitive awareness necessary for ball security. We expect the data to show that while distance matters, the timing and direction of travel are the primary drivers of impact on offensive effectiveness.

### NBA Official Team Game Logs & Advanced Statistics (nba_api) Dataset

This dataset provides team level, per game box score and advanced statistics for every NBA
regular season game across five seasons (2020-21 through 2024-25), fetched via the `nba_api`
Python library. The final cleaned dataset contains 12,000 team-game observations (30 teams,
~82 games per season × 5 seasons) across 68 columns. Each row represents one team's performance
in one game, meaning each game generates two rows, one for each team.

Data was pulled from two API endpoints. `TeamGameLogs` provides the per game box score data:
shooting, rebounds, turnovers, and point differential. `LeagueDashTeamStats` (in Advanced mode)
provides season aggregate metrics like Offensive Rating and Pace, which we joined onto each game
row as team quality baselines. From these, we engineered four additional variables that are
central to our research question: `eFG_PCT`, `TOV_PCT`, `REST_DAYS`, and `IS_B2B`.

**Key variables and what they mean:**

`FG_PCT` is raw field goal percentage - the fraction of shot attempts that go in. The league
average across our dataset is 0.470, with a standard deviation of 0.055. Individual game values
range from 0.277 to 0.689. Values below 0.40 represent poor shooting nights; above 0.55 is
excellent.

`eFG_PCT` (Effective Field Goal Percentage) = (FGM + 0.5 × FG3M) / FGA. It adjusts raw FG%
upward to account for the extra value of three-pointers, making it a more accurate measure of
shooting quality per possession. The dataset mean is 0.542 (std 0.066). The worst single game
performance in our data was a 0.301 (Houston, February 2021); the best was 0.804. This is our
**primary dependent variable**: it captures shooting efficiency per possession in a way that
controls for whether a team was taking more two or three point shots. The fatigue literature
predicts eastward travel and back-to-backs will push this value down.

`TOV_PCT` (Turnover Rate) = TOV / (FGA + 0.44 × FTA + TOV). This measures the fraction of
offensive possessions that end in a turnover rather than a shot. The dataset mean is 0.124
(std 0.034). Values above 0.18 indicate poor ball security. Cognitive fatigue from sleep
disruption is theorized to increase this rate on back-to-back road games. This is our **second
primary dependent variable**.

`REST_DAYS` is an engineered variable counting the number of full days between a team's current
and previous game, derived by differencing consecutive `GAME_DATE` values within each
team-season. It ranges from 0 to 12 days in our dataset. A value of 0 means the team played
the previous night — a back-to-back. Season openers are assigned 3 days (a neutral baseline).
`IS_B2B` is a binary indicator derived from `REST_DAYS == 0`; in our dataset, 2,153 games
(17.9%) are back-to-backs.

`SEASON_OFF_RATING` is a season-aggregate advanced stat measuring points scored per 100
possessions. It comes from `LeagueDashTeamStats` and ranges from 105.2 (Cleveland, 2020-21)
to 122.8 across team seasons. It serves as a team quality baseline: when a team that averages
120 OE scores at a 108 OE rate on a back-to-back road trip, that gap is the signal we are
trying to explain.

**Concerns with this dataset:**

The biggest concern is load management confounding. The 2020-25 era saw heavy use of
deliberate player resting, particularly for star players on road back to backs. A team's low
`eFG_PCT` on a travel night might partially reflect that their best shooter was sitting out
rather than fatigued. We cannot disentangle these effects without player availability data,
which is a known limitation.

A secondary concern is **home-court advantage confounding**. Visiting team shooting deficits
could reflect crowd noise, unfamiliar shot clocks, or referee bias in addition to travel fatigue.
Our models will include `IS_AWAY` as a direct control, but complete isolation of travel fatigue
from all home-court effects is not possible from this dataset alone.

Finally, the **2020-21 season** contains only 2,160 rows (72 games per team) versus 2,460 for
all other seasons due to the delayed COVID-era start. This season also features a more compressed
schedule with proportionally more back-to-backs. We retain it in the dataset but flag it as a
potential moderator in robustness checks.

### Baseline Model
Building on our exploratory findings, we implement a baseline regression model to formally estimate how fatigue-related variables impact offensive performance. By including controls for team, opponent, and season, this model isolates the effect of travel and scheduling factors beyond differences in team strength and game context.

```python
df = df.copy()
```

```python
import statsmodels.formula.api as smf

# List of predictors (fatigue variables)
predictors = [
    "REST_DAYS",
    "IS_B2B",
    "TRAVEL_MILES",
    "TZ_CROSSED",
    "TRAVEL_LAST_3",
    "IS_AWAY"
]

# Create formula for statsmodels
# Example for GAME_OE
formula = "GAME_OE ~ " + " + ".join(predictors) + " + C(TEAM) + C(SEASON) + C(OPPONENT)"

# Fit OLS model with clustered standard errors by TEAM_ID
model_game_oe = smf.ols(formula=formula, data=df).fit(
    cov_type="cluster",
    cov_kwds={"groups": df["TEAM"]}
)

# Summary
print(model_game_oe.summary())
```

The table above reports the results of our OLS regression, where each coefficient represents the estimated effect of a predictor on offensive efficiency while holding all other variables constant. The “coef” column shows the direction and magnitude of each effect, while the “P>|z|” column indicates whether the effect is statistically significant. In the following section, we focus on the key fatigue-related variables to interpret which factors meaningfully impact offensive performance.

```python
import pandas as pd
import statsmodels.formula.api as smf

# Fatigue predictors + control
predictors = ["REST_DAYS", "IS_B2B", "TRAVEL_MILES", "TZ_CROSSED", "TRAVEL_LAST_3", "IS_AWAY"]

# Outcomes
outcomes = ["GAME_OE", "eFG_PCT", "TOV_PCT"]

# Dictionary to store results
results_dict = {}

for outcome in outcomes:
    formula = outcome + " ~ " + " + ".join(predictors) + " + C(TEAM) + C(SEASON) + C(OPPONENT)"
    model = smf.ols(formula=formula, data=df).fit(
        cov_type="cluster",
        cov_kwds={"groups": df["TEAM"]}
    )
    results_dict[outcome] = model

# Create summary table
summary_table = pd.DataFrame()

for outcome, model in results_dict.items():
    coef_df = model.params.to_frame(name="coef").reset_index()
    coef_df = coef_df.rename(columns={"index": "variable"})
    coef_df["outcome"] = outcome
    coef_df["pval"] = model.pvalues.values
    summary_table = pd.concat([summary_table, coef_df], axis=0)

# Keep only the main predictors (optional, remove FE if desired)
summary_table = summary_table[summary_table["variable"].isin(predictors)]

# Pivot for readability
summary_table_pivot = summary_table.pivot(index="variable", columns="outcome", values="coef")
pvals_pivot = summary_table.pivot(index="variable", columns="outcome", values="pval")

print("=== Coefficients Table ===")
print(summary_table_pivot.round(3))
print("\n=== P-Values Table ===")
print(pvals_pivot.round(3))
```

The tables above summarize the regression results across multiple outcome variables. The coefficients table shows the estimated effect of each fatigue-related predictor on offensive efficiency, turnover percentage, and effective field goal percentage, while the p-values table indicates whether these effects are statistically significant. By comparing results across outcomes, we can identify whether fatigue variables impact different aspects of performance in distinct ways. In the following section, we focus on the most relevant predictors to interpret their effects in more detail.

### From this model, we observe that:
- REST_DAYS has a p-value of 0.868. The effect we get from this predictor is basically 0, not statistically significant. So, additional rest days do not measurably improve offensive efficiency once team and opponent strength are controlled.
- Back-to-back has a p-value of 0.011. This variable reduces offensive efficiency by a small, but statistically significant, amount.
- Travel distance has a p-value of 0.842, signifying that the effect is extremely small and not significant.
- Time Zones crossed has a p-value of 0.180, signifying that the effect is extremely small and not significant.
- Travel in last 3 games has a p-value of 0.260, signifying that the effect is extremely small and not significant.
- Away games predictor has a p-value less than 0.001. It is the largest and strongest effect in determining offensive efficiency among the chose predictors.  

### Interpretation of Results:
- The Effect of Fatigue: The most statistically significant drags on a team's offensive efficiency and shooting are playing on the road (IS_AWAY, p < 0.001) and playing back-to-back games (IS_B2B, p = 0.011 for GAME_OE and p = 0.004 for eFG_PCT). Being the away team costs teams about 1.73 points per 100 possessions, and playing on zero days of rest costs them a further 0.99 points.

- Mental vs. Physical Fatigue: A noticable diverge in the results occurs when you look at travel. Travel distance by itself does not have a large impact, but noticable for turnover percentage, crossing timezones (TZ_CROSSED, p=0.030) and heavy travel over a series of games (TRAVEL_LAST_3, p=0.022) have a noticable effect, increasing turnover percentage. One theory we can glean from this information is that the built up fatigue from travel doesn't come as a physical effect such as your body failing when trying to shoot, but as mental fatigue causing you to make bad decisions when passing or other less physically dependent tasks. This interpretation is consistent with our correlation analysis, which shows weak relationships between travel variables and shooting efficiency but a clearer association between travel-related variables and turnover percentage.

- Diminishing Returns on Rest: Rest days showed no statistical significance across any model. In the context of our work, this can be thought of as the fact that the real impact cause is having no rest days (playing back-to-back games), but once a team has had a day off, more days off after that doesn't suddenly boost a teams performance.

Overall, these results suggest that while certain scheduling factors such as back-to-back games and playing away have statistically significant effects on performance, the impact of travel-related variables is generally weak when considered individually. This indicates that fatigue does not strongly reduce offensive efficiency in a direct, uniform way. Instead, its effects appear to be more subtle and context-dependent, influencing specific aspects of performance (such as turnovers) rather than overall scoring efficiency. ## Discusison and Conclusion


Our results suggest that the relationship between travel fatigue and offensive performance is more selective than our original hypothesis predicted. While schedule-related factors such as playing away and back-to-back games show consistent negative effects on offensive performance, travel-related variables do not exhibit strong direct impacts on overall efficiency. This contrasts with our expectation that long-distance travel and time-zone changes would significantly reduce shooting performance, suggesting that the effects of travel may be less direct than initially hypothesized and instead operate through more specific channels such as increased turnovers.

Taken together, these findings expand on the background research by suggesting that the relationship between travel and performance is more subtle than a simple distance-based fatigue model. Instead of a direct “miles traveled → worse shooting” effect, the data suggest that contextual factors such as playing away from home and compressed schedules are stronger predictors of offensive decline. Travel may still contribute indirectly, particularly through mental fatigue accumulated over several games, but its effect appears smaller than originally expected.

There are several possible explanations for the contradictory or weak results regarding travel distance and time zones. First, NBA teams travel under extremely optimized conditions compared to typical travelers: charter flights, advanced recovery programs, sports science staff, and sleep management strategies. These resources likely reduce the physiological costs of long-distance travel. Second, teams may strategically rest players or adjust rotations during demanding travel stretches, which could dilute measurable effects at the team level. Finally, offensive efficiency in basketball is influenced by many factors—shot selection, defensive schemes, player injuries, and randomness—making it difficult for a small set of travel variables to explain large portions of the variance.

These conclusions should be interpreted with caution. The main regression model explains only a modest portion of the variation in offensive efficiency (R² = 0.079), indicating that many important factors influencing game-to-game offensive performance were not included in the analysis. Additionally, several fatigue-related predictors are correlated with one another, particularly travel miles, time zones crossed, and cumulative recent travel, which may introduce multicollinearity and reduce the precision of individual coefficient estimates. The dataset is also constructed at the team-game level, which may obscure player-level fatigue effects. For example, individual players may experience fatigue differently depending on minutes played, injuries, or recovery strategies, but these effects are not directly observable in team-level data.

Future research could improve on this analysis in several ways. One promising direction would be to incorporate player-level tracking data, including minutes played, player travel distance, and individual shooting metrics. Another improvement would be to model interaction effects more explicitly—such as the combination of eastward travel, short rest, and long road trips—to better capture the “fatigue tipping point” proposed in the hypothesis. Additionally, longitudinal fatigue models that track players or teams across entire road trips could reveal patterns that are not visible in single-game analyses.

Overall, our findings provide partial support for the idea that NBA schedule fatigue affects offensive performance. Rather than being driven by travel distance alone, performance declines appear to be more strongly associated with contextual scheduling factors such as playing away and compressed game schedules. Travel-related fatigue may still play a role, but its effects are more subtle and are better reflected in specific aspects of performance, such as turnovers, rather than overall scoring efficiency.
