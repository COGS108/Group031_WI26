# Research Question

Does cumulative travel stress, quantified by the interaction between total distance traveled (miles), time-zone shifts, and rest intervals (e.g., back to back games), predict a statistically significant decline in a team's Offensive Efficiency (OE) relative to their season average?

Specifically, can we develop a multiple linear regression model that accurately identifies the amount of fatigue tax on shooting percentages (teams FG%) and turnover rates for visiting teams?

This project will test whether geographic factors like eastward travel and high-mileage road trips correlate with specific performance drops. By isolating these variables, we aim to determine if a predictive threshold exists where travel fatigue becomes a dominant predictor of offensive variance.

## Background and Prior Work

The modern landscape of professional basketball is defined by a rigorous competitive structure that frequently pushes the boundaries of human physiological and cognitive limits.

Within the National Basketball Association (NBA), the standard 82-game regular season involves a high density of competition interspersed with frequent transmeridian travel, creating a unique environment where athletic performance is perpetually modulated by recovery status and circadian alignment.

Travel stress in this context is characterized by two distinct but overlapping phenomena: travel fatigue—the acute exhaustion from the logistics of movement—and jet lag, a circadian rhythm disorder caused by crossing multiple time zones. This biological asymmetry is critical, as the human circadian system is naturally more adept at adapting to westward travel (phase delay) than shortening its cycle through eastward travel (phase advance) <a name="cite_note-1"></a><sup>1</sup>.

Research indicates that sleep disruption is the primary mechanism through which travel stress erodes performance. On the first night after travel, athletes experience a predictable reduction in sleep duration based on the number of time zones crossed, with eastward travel being significantly more disruptive (averaging -24.5 minutes) than westward travel <a name="cite_note-2"></a><sup>2</sup>.

These disruptions can manifest as significant decrements in Offensive Efficiency (OE), a holistic metric that measures scoring effectiveness per possession. Specifically, studies have shown that eastward jet lag is associated with a 1.2% decrease in Effective Field Goal Percentage ($eFG\%$) differential <a name="cite_note-3"></a><sup>3</sup>.

Furthermore, mental fatigue disrupts psychomotor vigilance, essential for shooting accuracy and ball security, often leading to a rise in turnover rates (TOV%) at the end of long road trips.

Existing literature has also identified specific "tipping points" where travel fatigue becomes the dominant predictor of performance variance. One of the most robust findings is that cumulative time zone changes of three or more within a three-day period are significantly detrimental to performance <a name="cite_note-4"></a><sup>4</sup>.

Additionally, the back-to-back game scenario remains the most consistent predictor of decline; visiting teams playing on the second night of a back-to-back win only about 36% of the time.

While total mileage is a factor, its impact is most severe when combined with the duration of a multi-city tour, as the cumulative stress of changing environments leads to a measurable decline in $eFG\%$.

To capture these complex, non-linear interactions, modern sports analytics projects have increasingly moved from simple linear models toward advanced techniques like Random Forests and Gradient Boosting (LightGBM) to accurately identify and predict the "fatigue tax" <a name="cite_note-5"></a><sup>5</sup>.

# Hypothesis


We hypothesize that **cumulative travel stress**, more specifically the interaction of eastward travel (time zone shifts), total distance covered, and limited rest (playing back to back games), will serve as a **statistically significant predictor** of a decline in a visiting team’s Offensive Efficiency (OE).

### Specific Predictions
* **Directional Asymmetry:** Eastward travel across two or more time zones will result in a much more noticeable decline in Effective Field Goal Percentage ($eFG\%$) than westward travel of equal distance.
* **The "Fatigue Tax" Tipping Point:** The decline will follow a non-linear trend; the combination of high-mileage road trips and back-to-back scenarios creates a "tipping point" that significantly increases Turnover Rates ($TOV\%$).

### Reasoning
This prediction is based on the biological reality of **circadian asymmetry**, where the human body struggles more with "phase advance" (losing time traveling east) than "phase delay" (traveling west).



As highlighted in our background research, this sleep disruption impairs both the fine motor skills required for shooting accuracy and the cognitive awareness necessary for ball security. We expect the data to show that while distance matters, the timing and direction of travel are the primary drivers of impact on offensive effectiveness.

### Dataset 1: NBA Official Team Game Logs & Advanced Statistics (nba_api)

This dataset provides team level, per game box score and advanced statistics for every NBA
regular season game across five seasons (2020-21 through 2024-25), fetched via the `nba_api`
Python library. The final cleaned dataset contains 12,000 team-game observations (30 teams,
~82 games per season × 5 seasons) across 68 columns. Each row represents one team's performance
in one game, meaning each game generates two rows, one for each team.

Data was pulled from two API endpoints. `TeamGameLogs` provides the per game box score data:
shooting, rebounds, turnovers, and point differential. `LeagueDashTeamStats` (in Advanced mode)
provides season aggregate metrics like Offensive Rating and Pace, which we joined onto each game
row as team quality baselines. From these, we engineered four additional variables that are
central to our research question: `eFG_PCT`, `TOV_PCT`, `REST_DAYS`, and `IS_B2B`.

**Key variables and what they mean:**

`FG_PCT` is raw field goal percentage - the fraction of shot attempts that go in. The league
average across our dataset is 0.470, with a standard deviation of 0.055. Individual game values
range from 0.277 to 0.689. Values below 0.40 represent poor shooting nights; above 0.55 is
excellent.

`eFG_PCT` (Effective Field Goal Percentage) = (FGM + 0.5 × FG3M) / FGA. It adjusts raw FG%
upward to account for the extra value of three-pointers, making it a more accurate measure of
shooting quality per possession. The dataset mean is 0.542 (std 0.066). The worst single game
performance in our data was a 0.301 (Houston, February 2021); the best was 0.804. This is our
**primary dependent variable**: it captures shooting efficiency per possession in a way that
controls for whether a team was taking more two or three point shots. The fatigue literature
predicts eastward travel and back-to-backs will push this value down.

`TOV_PCT` (Turnover Rate) = TOV / (FGA + 0.44 × FTA + TOV). This measures the fraction of
offensive possessions that end in a turnover rather than a shot. The dataset mean is 0.124
(std 0.034). Values above 0.18 indicate poor ball security. Cognitive fatigue from sleep
disruption is theorized to increase this rate on back-to-back road games. This is our **second
primary dependent variable**.

`REST_DAYS` is an engineered variable counting the number of full days between a team's current
and previous game, derived by differencing consecutive `GAME_DATE` values within each
team-season. It ranges from 0 to 12 days in our dataset. A value of 0 means the team played
the previous night — a back-to-back. Season openers are assigned 3 days (a neutral baseline).
`IS_B2B` is a binary indicator derived from `REST_DAYS == 0`; in our dataset, 2,153 games
(17.9%) are back-to-backs.

`SEASON_OFF_RATING` is a season-aggregate advanced stat measuring points scored per 100
possessions. It comes from `LeagueDashTeamStats` and ranges from 105.2 (Cleveland, 2020-21)
to 122.8 across team seasons. It serves as a team quality baseline: when a team that averages
120 OE scores at a 108 OE rate on a back-to-back road trip, that gap is the signal we are
trying to explain.

**Concerns with this dataset:**

The biggest concern is load management confounding. The 2020-25 era saw heavy use of
deliberate player resting, particularly for star players on road back to backs. A team's low
`eFG_PCT` on a travel night might partially reflect that their best shooter was sitting out
rather than fatigued. We cannot disentangle these effects without player availability data,
which is a known limitation.

A secondary concern is **home-court advantage confounding**. Visiting team shooting deficits
could reflect crowd noise, unfamiliar shot clocks, or referee bias in addition to travel fatigue.
Our models will include `IS_AWAY` as a direct control, but complete isolation of travel fatigue
from all home-court effects is not possible from this dataset alone.

Finally, the **2020-21 season** contains only 2,160 rows (72 games per team) versus 2,460 for
all other seasons due to the delayed COVID-era start. This season also features a more compressed
schedule with proportionally more back-to-backs. We retain it in the dataset but flag it as a
potential moderator in robustness checks.

```python
# --- Imports & setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import os, time, warnings

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted')

from nba_api.stats.endpoints import TeamGameLogs, LeagueDashTeamStats

for path in ['data/00-raw', 'data/01-interim', 'data/02-processed']:
    os.makedirs(path, exist_ok=True)
```

```python
# --- Fetch & cache raw data ---
SEASONS = ['2020-21', '2021-22', '2022-23', '2023-24', '2024-25']
RAW_GAMELOGS_PATH  = 'data/00-raw/nba_team_gamelogs_raw.csv'
RAW_ADVANCED_PATH  = 'data/00-raw/nba_advanced_stats_raw.csv'

def fetch_gamelogs(seasons, delay=0.7):
    frames = []
    for s in seasons:
        df = TeamGameLogs(season_nullable=s, season_type_nullable='Regular Season').get_data_frames()[0]
        df['SEASON'] = s
        frames.append(df)
        time.sleep(delay)
    return pd.concat(frames, ignore_index=True)

def fetch_advanced(seasons, delay=0.7):
    frames = []
    for s in seasons:
        df = LeagueDashTeamStats(season=s, season_type_all_star='Regular Season',
                                 measure_type_detailed_defense='Advanced').get_data_frames()[0]
        df['SEASON'] = s
        frames.append(df)
        time.sleep(delay)
    return pd.concat(frames, ignore_index=True)

if not os.path.exists(RAW_GAMELOGS_PATH):
    fetch_gamelogs(SEASONS).to_csv(RAW_GAMELOGS_PATH, index=False)
if not os.path.exists(RAW_ADVANCED_PATH):
    fetch_advanced(SEASONS).to_csv(RAW_ADVANCED_PATH, index=False)

df_raw     = pd.read_csv(RAW_GAMELOGS_PATH)
df_adv_raw = pd.read_csv(RAW_ADVANCED_PATH)

print(f"Game logs : {df_raw.shape}")
print(f"Advanced  : {df_adv_raw.shape}")
df_raw.head(3)
```

```python
# --- Engineer variables & merge ---
df = df_raw.copy()
df['GAME_DATE'] = pd.to_datetime(df['GAME_DATE'])
df['IS_AWAY']   = df['MATCHUP'].str.contains('@').astype(int)
df['eFG_PCT']   = (df['FGM'] + 0.5 * df['FG3M']) / df['FGA']
df['TOV_PCT']   = df['TOV'] / (df['FGA'] + 0.44 * df['FTA'] + df['TOV'])
df['WIN']       = (df['WL'] == 'W').astype(int)

df = df.sort_values(['TEAM_ID', 'GAME_DATE']).reset_index(drop=True)
df['REST_DAYS'] = (df.groupby(['TEAM_ID', 'SEASON'])['GAME_DATE']
                     .diff().dt.days - 1).fillna(3).clip(lower=0).astype(int)
df['IS_B2B']    = (df['REST_DAYS'] == 0).astype(int)

df_adv = (df_adv_raw[['TEAM_ID', 'SEASON', 'OFF_RATING', 'DEF_RATING', 'NET_RATING', 'PACE']]
            .rename(columns={'OFF_RATING': 'SEASON_OFF_RATING', 'DEF_RATING': 'SEASON_DEF_RATING',
                             'NET_RATING': 'SEASON_NET_RATING', 'PACE': 'SEASON_PACE'}))
df = df.merge(df_adv, on=['TEAM_ID', 'SEASON'], how='left')

print(f"Shape: {df.shape}  |  Away: {df['IS_AWAY'].sum():,}  |  B2B: {df['IS_B2B'].sum():,} ({df['IS_B2B'].mean():.1%})")
print(f"Date range: {df['GAME_DATE'].min().date()} to {df['GAME_DATE'].max().date()}")
df[['TEAM_ABBREVIATION', 'GAME_DATE', 'MATCHUP', 'REST_DAYS', 'IS_B2B', 'eFG_PCT', 'TOV_PCT']].head(5)
df.to_csv("data/01-interim/engineered_features.csv", index=False)
```

```python
# --- Outlier detection (3xIQR) & distributions ---
def iqr_outliers(series, label, multiplier=3.0):
    Q1, Q3 = series.quantile(0.25), series.quantile(0.75)
    IQR = Q3 - Q1
    flagged = ((series < Q1 - multiplier*IQR) | (series > Q3 + multiplier*IQR)).sum()
    print(f"{label:12s}  range [{series.min():.3f}, {series.max():.3f}]  flagged: {flagged}")

print("Outlier check (3xIQR) — statistical outliers retained, only impossible values dropped:")
for col in ['eFG_PCT', 'TOV_PCT', 'FG_PCT', 'REST_DAYS']:
    iqr_outliers(df[col].astype(float), col)

print("\nNote: REST_DAYS flags many rows because the distribution peaks sharply at 1; "
      "all values (0–12) are valid and intentionally retained.")

# Distributions
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
for ax, (col, label, color) in zip(axes, [
    ('eFG_PCT', 'Effective FG%', '#4C72B0'),
    ('TOV_PCT', 'Turnover Rate', '#DD8452'),
    ('REST_DAYS','Rest Days',    '#C44E52')
]):
    ax.hist(df[col].dropna(), bins=40, color=color, edgecolor='white', alpha=0.85)
    ax.axvline(df[col].mean(), color='black', linestyle='--', lw=1.5,
               label=f'Mean: {df[col].mean():.3f}')
    ax.set_title(label); ax.legend(fontsize=9)
plt.suptitle('Key Variable Distributions', fontsize=13)
plt.tight_layout()
plt.savefig('data/01-interim/fig_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

### Dataset 2: NBA Schedule & Arena Location Data (JSON Schedule Files)

This dataset provides official NBA regular season schedule information for every NBA
regular season game across five seasons (2020-21 through 2024-25), parsed from structured
JSON schedule files. The final cleaned dataset contains approximately 8,000+ game
observations (30 teams, ~82 games per season, 5 seasons) across 9 columns.
Each row represents one NBA game, meaning each observation includes both a home
and away team.

Data was parsed from season-specific JSON files (`nba_schedule_<season>.json`).
For each month and game entry within the nested JSON structure, we extracted the
game date, home and away team names and abbreviations, arena name, and final scores.
After parsing, we filtered the dataset to include only official NBA franchises
(30 teams), ensuring removal of malformed or non-standard entries. This dataset
forms the structural backbone for constructing travel-related variables central
to our research question.

**Key variables and what they mean:**

`season` identifies the NBA season (e.g., 2020 corresponds to the 2020-21 season).
This allows us to group games within team-season sequences for chronological travel
calculations.

`date` is the official game date, converted to datetime format. This variable is
essential for computing inter-game rest intervals and sequencing consecutive
games for each team.

`home_team` and `away_team` are full franchise names. `home_abbr` and
`away_abbr` provide standardized team abbreviations that enable merging
with team-level performance data.

`arena` identifies the venue where the game was played. This variable is
critical because it allows us to merge in geographic coordinates (latitude
and longitude) from a separate arena-location dataset. These coordinates
enable calculation of travel distance between consecutive game locations.

`home_score` and `away_score` represent final game outcomes and allow calculation
of game margin (`home_score − away_score`), which can later be used to measure
performance relative to expectation.

This dataset enables construction of the independent variables measuring
cumulative travel stress. Using chronological sorting within each team-season,
we can compute location transitions, road trip length (consecutive away games),
travel distance between arenas, and time-zone shifts between consecutive games.
These engineered variables allow us to operationalize geographic and circadian
stress, which we hypothesize predicts declines in offensive efficiency.

**Concerns with this dataset:**

The primary limitation is that arena names do not inherently include geographic
coordinates; latitude, longitude, and time zone data must be merged from an
external source. Travel distances are therefore approximations based on direct
arena-to-arena displacement and do not reflect actual charter flight routes or
delays.

A secondary concern is the absence of game start time, which restricts our
ability to directly measure circadian misalignment (e.g., body-clock disadvantage
following eastward travel).

Finally, the **2020-21 season** contains fewer games per team (72 instead of 82)
due to the delayed COVID-era start. This season also features a more compressed
schedule with proportionally more back-to-backs. We retain it in the dataset
but flag it as a potential moderator in robustness checks.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import json
import math
import pandas as pd
import numpy as np

# Load/parse each season
seasons = [2020, 2021, 2022, 2023, 2024, 2025]
all_games = []

for season in seasons:
    try:
        with open(f"data/00-raw/nba_schedule_{season}.json") as f:
            data = json.load(f)
        
        for month in data['lscd']:
            for game in month['mscd']['g']:
                all_games.append({
                    'season':     season,
                    'date':       game['gdte'],
                    'home_team':  game['h']['tc'] + ' ' + game['h']['tn'],
                    'away_team':  game['v']['tc'] + ' ' + game['v']['tn'],
                    'home_abbr':  game['h']['ta'],
                    'away_abbr':  game['v']['ta'],
                    'arena':      game.get('an', ''),
                    'home_score': game['h'].get('s', None),
                    'away_score': game['v'].get('s', None),
                })
    except FileNotFoundError:
        print(f"Season {season} not found — skipping")
        
# Build df
schedule = pd.DataFrame(all_games)

# Fix data types
schedule['date']       = pd.to_datetime(schedule['date'])
schedule['home_score'] = pd.to_numeric(schedule['home_score'], errors='coerce')
schedule['away_score'] = pd.to_numeric(schedule['away_score'], errors='coerce')

# Filter to only real NBA teams
NBA_TEAMS = {
    "Atlanta Hawks", "Boston Celtics", "Brooklyn Nets", "Charlotte Hornets",
    "Chicago Bulls", "Cleveland Cavaliers", "Dallas Mavericks", "Denver Nuggets",
    "Detroit Pistons", "Golden State Warriors", "Houston Rockets", "Indiana Pacers",
    "LA Clippers", "Los Angeles Lakers", "Memphis Grizzlies", "Miami Heat",
    "Milwaukee Bucks", "Minnesota Timberwolves", "New Orleans Pelicans",
    "New York Knicks", "Oklahoma City Thunder", "Orlando Magic",
    "Philadelphia 76ers", "Phoenix Suns", "Portland Trail Blazers",
    "Sacramento Kings", "San Antonio Spurs", "Toronto Raptors",
    "Utah Jazz", "Washington Wizards"
}

schedule_clean = schedule[
    schedule['home_team'].isin(NBA_TEAMS) &
    schedule['away_team'].isin(NBA_TEAMS)
].reset_index(drop=True)

# Summary
print(f"Seasons covered: {sorted(schedule_clean['season'].unique())}")
print(f"Date range: {schedule_clean['date'].min()} to {schedule_clean['date'].max()}")
print(f"Shape: {schedule_clean.shape}")
print(schedule_clean.head())

df.to_csv("data/02-processed/schedule_clean.csv", index=False)
```

```python
team_df['prev_lat'] = team_df.groupby('TEAM')['lat'].shift(1)
team_df['prev_lon'] = team_df.groupby('TEAM')['lon'].shift(1)
team_df['prev_tz']  = team_df.groupby('TEAM')['tz'].shift(1)
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 3958.8  # miles
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

team_df['TRAVEL_MILES'] = team_df.apply(
    lambda row: haversine(row.prev_lat, row.prev_lon,
                          row.lat, row.lon)
    if pd.notnull(row.prev_lat) else 0,
    axis=1
)
```

```python
timezones = {
    "Eastern": 0,
    "Central": 1,
    "Mountain": 2,
    "Pacific": 3
}
team_df["tz_num"] = team_df["tz"].map(timezones)
team_df['prev_tz_num'] = team_df.groupby('TEAM')['tz_num'].shift(1)
team_df["TZ_CROSSED"] = (team_df['tz_num'] - team_df['prev_tz_num']).abs().fillna(0)
team_df['TRAVEL_LAST_3'] = (
    team_df.groupby('TEAM')['TRAVEL_MILES']
           .rolling(3, min_periods=1)
           .sum()
           .reset_index(0, drop=True)
)
combined_df = team_df.merge(
    cleaned_gamelogs, 
    how="left", 
    left_on=["date", "TEAM", "OPPONENT"],
    right_on=["GAME_DATE", "TEAM_NAME", "MATCHUP"]    
)
cleaned_df = combined_df.dropna()
cleaned_df = cleaned_df[['SEASON', 'TEAM', 'OPPONENT', 'TRAVEL_MILES', 'TZ_CROSSED', 
                         'REST_DAYS', 'IS_B2B', 'TRAVEL_LAST_3', 'IS_AWAY', 'GAME_OE', 
                         'FG_PCT', 'eFG_PCT', 'TOV_PCT']]
cleaned_df.to_csv("data/02-processed/cleaned_df.csv", index=False)
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# Fatigue predictors + control
predictors = ["REST_DAYS", "IS_B2B", "TRAVEL_MILES", "TZ_CROSSED", "TRAVEL_LAST_3", "IS_AWAY"]

# Outcomes
outcomes = ["GAME_OE", "eFG_PCT", "TOV_PCT"]

# Dictionary to store results
results_dict = {}

for outcome in outcomes:
    formula = outcome + " ~ " + " + ".join(predictors) + " + C(TEAM) + C(SEASON) + C(OPPONENT)"
    model = smf.ols(formula=formula, data=df).fit(
        cov_type="cluster",
        cov_kwds={"groups": df["TEAM"]}
    )
    results_dict[outcome] = model

# Create summary table
summary_table = pd.DataFrame()

for outcome, model in results_dict.items():
    coef_df = model.params.to_frame(name="coef").reset_index()
    coef_df = coef_df.rename(columns={"index": "variable"})
    coef_df["outcome"] = outcome
    coef_df["pval"] = model.pvalues.values
    summary_table = pd.concat([summary_table, coef_df], axis=0)

# Keep only the main predictors (optional, remove FE if desired)
summary_table = summary_table[summary_table["variable"].isin(predictors)]

# Pivot for readability
summary_table_pivot = summary_table.pivot(index="variable", columns="outcome", values="coef")
pvals_pivot = summary_table.pivot(index="variable", columns="outcome", values="pval")

print("=== Coefficients Table ===")
print(summary_table_pivot.round(3))
print("\n=== P-Values Table ===")
print(pvals_pivot.round(3))
```
